```python
import csv
import os

def csv_to_list(folder):
    all_data = []

    for filename in os.listdir(folder):
        if not filename.endswith("_normalized.csv"):
            continue

        filepath = os.path.join(folder, filename)
        print(f"ğŸ” íŒŒì¼ ì½ëŠ” ì¤‘: {filepath}")

        with open(filepath, "r", encoding="utf-8-sig") as f:
            reader = csv.reader(f)
            header = next(reader)  # í—¤ë” ìŠ¤í‚µ

            for row in reader:
                # 1) ë¹ˆ ë¬¸ìì—´ ì œê±°
                row = [c.strip() for c in row if c.strip() != ""]

                # 2) ìµœì†Œ 6ì¹¸ì¸ì§€ í™•ì¸
                if len(row) < 6:
                    continue

                # 3) BOM ì œê±°
                row[0] = row[0].replace("\ufeff", "")

                # 4) ê°’ ë³€í™˜
                try:
                    rank = int(row[0])
                    model = row[1]
                    sales = int(row[2])
                    share = float(row[3])
                    mom = row[4]
                    yoy = row[5]
                except Exception as e:
                    print("âš  ë°ì´í„° ë³€í™˜ ì˜¤ë¥˜:", row, e)
                    continue

                all_data.append([
                    rank, model, sales, share, mom, yoy
                ])

    return all_data

kia_list = csv_to_list("kia")
hyundai_list = csv_to_list("hyundai")

print(kia_list[:5])
print(hyundai_list[:5])
```
