```python
# 3.
import pandas as pd
import glob
import os

folder_name = 'KIA'
# CSV 파일들이 들어있는 폴더 경로
folder_path = f"./{folder_name}"

# 폴더 안의 모든 CSV 파일 가져오기
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

merged_data = []
category_line = None
model_line = None

for idx, file in enumerate(csv_files):
    # CSV 전체를 줄 단위로 읽기
    with open(file, "r", encoding="utf-8") as f:
        lines = f.read().splitlines()

    # 각 파일의 첫 두 줄은 헤더 형식(카테고리 / 모델명)
    if idx == 0:
        # 첫 파일의 헤더 2줄만 저장
        category_line = lines[0]
        model_line = lines[1]

    # 3번째 줄부터는 실제 주간 데이터
    data_lines = lines[2:]

    # 병합 리스트에 추가
    merged_data.extend(data_lines)

# 최종 파일 생성
file_name = 'KIAall'
output_file = f"{file_name}.csv"
with open(output_file, "w", encoding="utf-8") as f:
    f.write(category_line + "\n")
    f.write(model_line + "\n")
    for line in merged_data:
        f.write(line + "\n")

print("병합 완료:", output_file)
```

```python
#4. 병합 최종 // folder_name: csv 파일들이 들어있는 경로, file_name: 출력될 파일명만 수정하여 사용
import pandas as pd
import glob
import os

folder_name = 'HYUNDAI' # 입력할 폴더명 수정
# CSV 파일들이 들어있는 폴더 경로
folder_path = f"./{folder_name}"
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

dfs = []

for file in csv_files:
    with open(file, "r", encoding="utf-8") as f:
        lines = [line.strip() for line in f.readlines() if line.strip()]

    # 1번 줄: 카테고리 → 사용하지 않음
    # 2번 줄: 헤더
    header_line = lines[1]

    # 쉼표 기준 split (정상적으로 6개 컬럼이 나와야 함)
    header = [h.strip() for h in header_line.split(",")]

    # 데이터 줄들 split
    data_rows = [line.split(",") for line in lines[2:]]

    # DataFrame 생성
    df = pd.DataFrame(data_rows, columns=header)

    # 날짜 컬럼 이름 통일
    df.rename(columns={header[0]: "주"}, inplace=True)

    dfs.append(df)

# 첫 번째 DF를 기준으로 가로 병합
merged_df = dfs[0]

for df in dfs[1:]:
    merged_df = pd.merge(merged_df, df, on="주", how="outer")

# 최종 헤더 구성
category_line = "카테고리: 자동차"
header_line = ",".join(merged_df.columns)

# 저장
file_name = 'HYUNDAIall' # 출력될 파일명 수정

output_file = f"{file_name}.csv"
with open(output_file, "w", encoding="utf-8") as f:
    f.write(category_line + "\n")
    f.write(header_line + "\n")
    merged_df.to_csv(f, index=False, header=False)

print("가로 병합 완료:", output_file)
```
